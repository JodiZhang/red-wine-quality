### Background on Wine
Wine is the result of complete or partial fermentation of fresh grapes or of grape. By this definition, the quality of wine is closely related to the composition and variety of grape. Wines can also be distinguished by the geographic location of vineyards, viticultural practices and techniques used. The flavor profile comes from a balanced mixture of several chemical substances of different origin and structure, such as: acids, alcohols, inorganic substances, nitrogenous compounds, phenolics, water, etc. 

Wine can be classified as red, white and rosé  based on the level of sweetness, alcohol content, carbon dioxide content, color, grape variety, fermentation method and maturation process. Red wines are are a result of the alcoholic fermentation of solid grapes (skins and seed included) while white wines are produced by fermentation of grape juice.

We will be using Multiple Linear Regression to predict wine quality and identify the effects each feature has.

<b>Data Set Information:</b>

1. <b>Fixed Acidity</b>: most acids involved with wine or fixed or nonvolatile (do not evaporate readily)
2. <b>Volatile Acidity</b> : the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste
3. <b>Citric Acid</b> : found in small quantities, citric acid can add 'freshness' and flavor to wines
4. <b>Residual Sugar</b> : the amount of sugar remaining after fermentation stops, it's rare to find wines with less than 1 gram/liter and wines with greater than 45 grams/liter are considered sweet
5. <b>Chlorides</b> : the amount of salt in the wine
6. <b>Free Sulfur Dioxide</b> : the free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion; it prevents microbial growth and the oxidation of wine
7. <b>Total Sulfur Dioxideamount</b> : of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine, but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine
8. <b>Density</b> : the density of water is close to that of water depending on the percent alcohol and sugar content
9. <b>pH</b> : describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale
10. <b>Sulphates</b> : a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant
11. <b>Alcohol</b> : the percent alcohol content of the wine
12. <b>Quality</b> : output variable (based on sensory data, score between 0 and 10)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = wine[['alcohol','sulphates','citric_acid','fixed_acidity','residual_sugar']]
y = wine['quality']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=29)
```

```python
Model = LinearRegression()
Model.fit(X_train,y_train)
y_prediction = Model.predict(X_test)
```

```python
accuracy = Model.score(X_test,y_test)
print(accuracy)
```

```python
rsme = mean_squared_error(y_test,y_prediction)
print(rsme)
```
